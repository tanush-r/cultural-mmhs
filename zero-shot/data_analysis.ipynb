# Data Analysis

## Homophobia
## 1) English

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load data
train = pd.read_excel("homophobia/english/Train_labels.xlsx")
test  = pd.read_excel("homophobia/english/Test_labels.xlsx")

# Count labels
train_counts = train["label"].value_counts()
test_counts  = test["label"].value_counts()

# Align categories
categories = sorted(set(train_counts.index) | set(test_counts.index))
train_vals = [train_counts.get(c, 0) for c in categories]
test_vals  = [test_counts.get(c, 0) for c in categories]
```

```python
x = np.arange(len(categories))
w = 0.35

plt.figure(figsize=(8,5))
plt.bar(x - w/2, train_vals, width=w, label="Train")
plt.bar(x + w/2, test_vals,  width=w, label="Test")

plt.xticks(x, categories)
plt.ylabel("Count")
plt.title("Label Counts – Train vs Test")
plt.legend()
plt.grid(axis="y", alpha=0.3)

plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(train_vals, labels=categories, autopct="%1.1f%%", startangle=90)
plt.title(f"Train Split (n={sum(train_vals)})")
plt.show()
```

# Misogyny

```python
# Load data
train = pd.read_csv("misogyny/tamil/train.csv")
test  = pd.read_csv("misogyny/tamil/test_with_labels.csv")

# Count labels
train_counts = train["labels"].str.lower().value_counts()
test_counts  = test["labels"].str.lower().value_counts()

# Align categories
categories = sorted(set(train_counts.index) | set(test_counts.index))
train_vals = [train_counts.get(c, 0) for c in categories]
test_vals  = [test_counts.get(c, 0) for c in categories]
```

# Multimedia Automatic Misogyny Identification(MAMI)

```python
train = pd.read_csv("mami/train.tsv", sep="\t")
test = pd.read_csv("mami/test.tsv", sep="\t")
val = pd.read_csv("mami/validation.tsv", sep="\t")

# Binary label columns
label_cols = ["label", "shaming", "stereotype", "objectification", "violence"]
```

```python
def get_split_counts(df, columns):
    counts = {}
    for col in columns:
        counts[col] = df[col].value_counts().sort_index()
    return counts

train_counts = get_split_counts(train, label_cols)
test_counts  = get_split_counts(test, label_cols)
val_counts   = get_split_counts(val, label_cols)
```

```python
col = "label"

categories = [0, 1]
train_vals = [train_counts[col].get(c, 0) for c in categories]
test_vals  = [test_counts[col].get(c, 0)  for c in categories]
val_vals   = [val_counts[col].get(c, 0)   for c in categories]

x = np.arange(len(categories))
w = 0.25

plt.figure(figsize=(7,5))
plt.bar(x - w, train_vals, width=w, label="Train")
plt.bar(x,     test_vals,  width=w, label="Test")
plt.bar(x + w, val_vals,   width=w, label="Validation")

plt.xticks(x, ["0", "1"])
plt.ylabel("Count")
plt.title(f"{col.capitalize()} Distribution")
plt.legend()
plt.grid(axis="y", alpha=0.3)

plt.show()
```

```python
train_counts = [train[col].sum() for col in label_cols]
test_counts  = [test[col].sum() for col in label_cols]
val_counts   = [val[col].sum() for col in label_cols]

x = np.arange(len(label_cols))
w = 0.25

plt.figure(figsize=(10,5))

plt.bar(x - w, train_counts, width=w, label="Train")
plt.bar(x,     test_counts,  width=w, label="Test")
plt.bar(x + w, val_counts,   width=w, label="Validation")

plt.xticks(x, label_cols, rotation=30)
plt.ylabel("Number of Positive Samples")
plt.title("Label Distribution Across Splits")
plt.legend()
plt.grid(axis="y", alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

label_cols = ["label", "shaming", "stereotype",
              "objectification", "violence"]

fig, axes = plt.subplots(2, 3, figsize=(12, 8))
axes = axes.flatten()

for i, col in enumerate(label_cols):
    counts = train[col].value_counts().sort_index()
    vals = [counts.get(0, 0), counts.get(1, 0)]
    
    wedges, texts, autotexts = axes[i].pie(
        vals,
        labels=["0", "1"],
        autopct="%1.1f%%",
        startangle=90,
        pctdistance=0.65,
        labeldistance=1.1
    )
    
    for autotext in autotexts:
        autotext.set_fontsize(8)
    
    for text in texts:
        text.set_fontsize(9)

    axes[i].set_title(col.capitalize())

# Remove unused subplot (since we have 5 classes but 6 grid spots)
axes[-1].axis("off")

plt.suptitle("Train Dataset – Class Distribution (0 vs 1)", fontsize=14)
plt.tight_layout()
plt.show()
```
