# Data Analysis

## Homophobia
## 1) English

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load data
train = pd.read_excel("homophobia/english/Train_labels.xlsx")
test  = pd.read_excel("homophobia/english/Test_labels.xlsx")

# Count labels
train_counts = train["label"].value_counts()
test_counts  = test["label"].value_counts()

# Align categories
categories = sorted(set(train_counts.index) | set(test_counts.index))
train_vals = [train_counts.get(c, 0) for c in categories]
test_vals  = [test_counts.get(c, 0) for c in categories]
```

```python
x = np.arange(len(categories))
w = 0.35

plt.figure(figsize=(8,5))
plt.bar(x - w/2, train_vals, width=w, label="Train")
plt.bar(x + w/2, test_vals,  width=w, label="Test")

plt.xticks(x, categories)
plt.ylabel("Count")
plt.title("Label Counts – Train vs Test")
plt.legend()
plt.grid(axis="y", alpha=0.3)

plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(train_vals, labels=categories, autopct="%1.1f%%", startangle=90)
plt.title(f"Train Split (n={sum(train_vals)})")
plt.show()
```

# Misogyny
# 1) Tamil

```python
# Load data
train = pd.read_csv("misogyny/tamil/train.csv")
test  = pd.read_csv("misogyny/tamil/test_with_labels.csv")

# Count labels
train_counts = train["labels"].str.lower().value_counts()
test_counts  = test["labels"].str.lower().value_counts()

# Align categories
categories = sorted(set(train_counts.index) | set(test_counts.index))
train_vals = [train_counts.get(c, 0) for c in categories]
test_vals  = [test_counts.get(c, 0) for c in categories]
```

```python
x = np.arange(len(categories))
w = 0.35

plt.figure(figsize=(8,5))
plt.bar(x - w/2, train_vals, width=w, label="Train")
plt.bar(x + w/2, test_vals,  width=w, label="Test")

plt.xticks(x, categories)
plt.ylabel("Count")
plt.title("Label Counts – Train vs Test")
plt.legend()
plt.grid(axis="y", alpha=0.3)

plt.show()
```

```python
plt.figure(figsize=(7, 4))
plt.barh(categories, train_vals)
for i, v in enumerate(train_vals):
    plt.text(v + 0.5, i, f"{v} ({v/sum(train_vals)*100:.1f}%)", va="center")
plt.xlabel("Count")
plt.title(f"Train Split (n={sum(train_vals)})")
plt.grid(axis="x", alpha=0.3)
plt.tight_layout()
plt.show()
```
